#Aplicação do modelo de Machine Learnig

```python
from sklearn.cluster import KMeans
```

```python
# APlicando metodo cotovelo
def calcular_wcss(dados_perfil):
  wcss = []
  for K in range(1,11):
         kmeans_dados = KMeans(n_clusters = K, random_state=0, init = 'k-means++')
         kmeans_dados.fit(X=dados_perfil)
         wcss.append(kmeans_dados.inertia_)
  return wcss
```

```python
dados_perfil = perfil_geral[['peso','Triglicerídeos']]
dados_perfil.head()
```

```python
wcss_perfil = calcular_wcss(dados_perfil)
```

```python
#visualizando
for i in range(len(wcss_perfil)):
  print(f'o CLuster {i} possui valor de WCSS de: {wcss_perfil[i]}')
```

# visualizando o gráfico cotovelo

```python
# grafico do cotovelo
grafico_wcss= px.line( x = range(1,11),
                      y=wcss_perfil
                      )
fig = go.Figure(grafico_wcss)

fig.update_layout(title = 'calculando wcss',
                  title_x= 0.5,
                  xaxis_title = 'Número de Clusters',
                  yaxis_title = 'Valor do WCSS',
                  template = 'plotly_white'
                  )
fig.show()
```

# Criando agrupamento para cada paciente

```python
# Aplicando algoritmo K-means
kmeans_pacientes = KMeans(n_clusters=3, random_state=0, init = 'k-means++')
perfil_geral['Cluster'] = kmeans_pacientes.fit_predict(dados_perfil)
```

```python
perfil_geral.head()
```

#Calculando Centroides de cada cluster //  visualizando gráfico

```python
centroide_cluesters = kmeans_pacientes.cluster_centers_
centroide_cluesters
```

```python
# Função gera grafico
def gera_grafico(x,y,xlabel,y_label):
  grafico = px.scatter( x = x,
                      y = y,
                     color = perfil_geral['Cluster'])
  fig = go.Figure(grafico)
  fig.update_layout(title='Análise de Clusters',
                  xaxis_title= xlabel,
                  yaxis_title= y_label
                  )
  fig.show()
```

```python
# gerando grafico
gera_grafico(perfil_geral['peso'], perfil_geral['Triglicerídeos'], 'peso', 'Triglicerídeos')
```

#Analisando dados gerados

```python
x = 'Triglicerídeos'
y = 'peso'

gera_grafico(perfil_geral[x], perfil_geral[y],x,y)
```
